```python
row1 = ['*','*','*']
row2 = ['*','*','*']
row3 = ['*','*','*']
map = [row1, row2, row3]
position = input("Where do you want to put the treasure?")
column = int(position[0]) 
row = int(position[1])
map[column - 1][row - 1] = 'X'
print(f"{row1}\n{row2}\n{row3}")
```

```python
import random
user = int(input("Rock, Paper or scissors? type 1 2 or 3"))
print(user)
computer = random.randint(1,3)
print(computer)
if user == computer:
    print("draw")

if (user == 1 and computer == 2) or (user == 2 and computer == 1) or (user == 3 and computer == 2):
    print("you win")

if (computer == 1 and user == 2) or (computer == 2 and user == 1) or (computer == 3 and user == 2):
    print("computer wins")

```

```python
import random

your_pick = (input("Type 1 for Rock, 2 for Paper, 3 for scissors"))
if int(your_pick) >= 4:
    print("invalid input, You lose")
else:
    comp_pick = str(random.randint(1,3))
    conc = int(your_pick + comp_pick)
    print(conc)

    if your_pick == "1":
        your_pick = "Rock"
    elif your_pick == "2":
        your_pick = "Paper"
    elif your_pick == "3":
        your_pick = "Scissors"

    print(f"you picked {your_pick}")

    if comp_pick == "1":
        comp_pick = "Rock"
    elif comp_pick == "2":
        comp_pick = "Paper"
    elif comp_pick == "3":
        comp_pick = "Scissors"

    print(f"computer picked {comp_pick}")

    if comp_pick == your_pick:
        print("Draw")

    if conc == 13 or conc == 21 or conc == 32:
        print("You win! :) ")
    if conc == 12 or conc == 23 or conc == 31:
        print("You lose :( ")

```
